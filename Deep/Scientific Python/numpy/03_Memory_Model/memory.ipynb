### Copy and View


Numpy me ek atributes hai `.base` jo agr ye None diya to onrignal array hai agr khuch aur dikhyaya to yaha koi aur khuch hai

```python
a = np.array([1,2,3,4,5],dtype='int64')
v = a[0:2]
c = a.copy()
print(v.base) # iska mtlb V me jo element ha waha A ke hai 
print(c.base) # None iska mtlb c khuhd ka malik nhai hai
print(a.nbytes) # 5 element * 8 = 40 bytes

```

```python
array = np.array([1,2,3,4],dtype='int32')
print(array.nbytes) # 4 elements * 4 = 16  ( int32 ka size 4 btye hota hai)
```

```python
arrays = np.arange(1,1000,dtype='int64')
print(arrays.nbytes)

arrays = np.arange(1,1000,dtype='int32')
print(arrays.nbytes)


arrays = np.arange(1,1000,dtype='int8')
print(arrays.nbytes)
```
