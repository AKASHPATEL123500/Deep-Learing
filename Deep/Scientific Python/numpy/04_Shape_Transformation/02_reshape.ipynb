### Revel()

```python
orginal_array = np.arange(1,17)
print("Orignal array : ", orginal_array)

## convert into reshape
reshape_array = orginal_array.reshape(4,-1)
print("resphe array and auto col: \n",reshape)


## iss reshpe array ko wapas array me convert karna hai
again_convert_into_orginal_array = reshape_array.ravel()
print("again convert into orginal array or 1d : ",again_convert_into_orginal_array)

## ravel() and flatten() is same 
## ravel() = orginal se link hota hai
## flatten() = copy banata hai 

## falting araay 
falting_array = reshape_array.flatten()
print("falten array : ",falting_array)



```

### Tranpose ( .T)

*row to col and col to row thats it*

```python
orginal_array = np.arange(1,13)
print("orignal array : ",orginal_array)

## reshape this array into 3*4
reshape_array = orginal_array.reshape(3,4)
print("reshaped array : ",reshape_array)

print("array shpe after transpose : ",reshape_array.shape)


## trnaspose this array 
transpose = reshape_array.T
print("\n Transpose array: \n",transpose)
print("array shpe before transpose : ",transpose.shape)
```

```python
array = np.arange(1,5).reshape(2,2)
print("array : ",array)
print(array.shape)
print(array.ndim)

ravel = array.ravel()
print("Ravel : \n",ravel)
ravel[2] = 20
print("value upade : ",ravel) ## value upade in ravle
print(array) ## orginal array me bhi value upade hua hai revel se


flatten_array = array.flatten()
print("faltten arrayy : ",flatten_array)
flatten_array[2] = 3
print("faltten value upade : ",flatten_array) 
print(array)

## faltten()
## faltten orignal array ka copy kar leta hai
## and value chnage karne per koi farqk nahi padta hai orginal array per

## ravel()
## orginal array ki value o deta hai
## yaha orginal array ko copy nahi karta hai
## and issme value upade karne per 
## orginal array me bhi upade ho jata hai


```
