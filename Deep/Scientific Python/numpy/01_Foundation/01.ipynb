### 2. shape , ndim , size in NumPy

*Concept:*

`shape` → dimensions

`ndim` → number of axes

`size` → total elements

```python
arrays = np.array([[1,2,3],[4,5,6],[2,1,2]])  
print("Array  : \n", arrays)

print("array row and col : ",arrays.shape)  # row and col batata hai 
print("array  demensial : ", arrays.ndim)  # kitne dem ka ye matrixs hai [[ ]] 2d hai agr three baracets hota to [[[ ]]] 3D dem ka hota 
print("array size : ",arrays.size) # yaha array me kitne element hai ye bata hai
```

### 5. array_indexing_slicing

```python
arra = np.random.randint(0,10,(3,3))
print("array matrixses : \n", arra)

print("\n")
arrays = np.array([[9,7,8],
 [7,6,5],
 [9,9 ,8]])

# slicing indexing
# rows
# syntax : array[row index]
print("row 0 print : ",arrays[0])  # ouput : row0 : [9,7,8]
print("Row 1 print",arrays[1])  # ouput : row1 : [7,6,5]
print("row 2 print : ",arrays[2])  # ouput : row2 : [9,9,8]
print("row 0 ki index 1 value : ",arrays[0,1])  # ouput : row0 : [7]
print("Negative slicing row ke last index ke index 2 ki value",arrays[-1,2]) # output : row2: [8]  negative silicing indexing
print("Row 2 ke index 2 ki value : ",arrays[2,2]) # output : row2: [8]
print("row2 ke all elements : ",arrays[2,0:3]) # output : row2: [9,9,8]
print("all values",arrays[:,1:3])
print("\n")

# cols
# syntax : array[row all :,col index]
print(arrays[:,0])  # ouput : col0 : [9,7,9]
print(arrays[:,1])  # ouput : col1:  [7,6,9]
print(arrays[:,2])  # ouput : col2 : [8,5,8]
print(arrays[:,-1]) # last element print
print(arrays[:,-1])


# problem : Maan lo tumhe upar wale array ki pehli do rows aur aakhri do columns chahiye.
print("sub array or matrixs : \n",arrays[0:2,0:2])
print("sub array 2 or matrixs : \n",arrays[1:3,1:3])
print("sub array 3 or matrixs : \n",arrays[:,0:2])
```

# 7. Array iteration

#### 1. Basic iteration (in the python way)

```python
arr = np.arange(1,5).reshape(2,2)

# simple ara loop
for i in arr:
    print(i)
    
# nditer loop in arr
for i in np.nditer(arr):
    print(i,end='')

print("\n")

# op_flasg
op_flasg_array = np.array([1,2,3,4])
for i in np.nditer(op_flasg_array,op_flags=['readwrite']):
    i[...] = i * i

print(op_flasg_array)


# neumurate()
enumerate_array =  np.array([[1,2,3,4],[2,3,4,5]])
for idx, key in np.ndenumerate(enumerate_array):
    print(f"index: {idx }, value:{key}")
```

# 7.Copy vs View
