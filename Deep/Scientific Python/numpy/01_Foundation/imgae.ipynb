```python

import numpy as np
import matplotlib.pyplot as ptl
```

```python
image = np.zeros((10,10),dtype='int')
image[5,:] = 255
image[:,5] = 255
image[0:2,0:2] = 150
ptl.title("Rnadom Image genaret")
ptl.imshow(image,cmap='gray')
ptl.show()
```

```python
image2 = np.zeros((10,10), dtype='int')
print(image2[8])
image2[5] = 255
image2[:,5] = 255
image2[0:2,0:2] = 150
image2[8] = 150
ptl.title("Rnadom Image genaret")
ptl.imshow(image2,cmap='gray')
ptl.show()
```
