```python
class UserData:
    def __init__(self,firstName,lastName):
        self.firstName = firstName
        self.lastName = lastName
    
    def full_name(self):
        return f"{self.firstName} {self.lastName}"    
    
    def add_data(self,age):
        self.age = age
        
test1 = UserData("Akash","Reddy")
print("Test 1 print user fisrt name is : ",test1.firstName)
print("Test 1 print user last name is : ",test1.lastName)
print("User full name print : ", test1.full_name())


print("\n")

class Student:
    def __init__(self,roll_no,name,age):
        self.roll_no = roll_no
        self.name = name
        self.age = age
    
    def show(self):
        print("Roll No : ",self.roll_no) 
        print("Name : ",self.name)
        print("Age : ",self.age)
         


s1 = Student(1,"Akash Reddy",16)
# print(s1.roll_no)
# print(s1.name)
# print(s1.age)
s1.show()

print("\n")

# inerhtence
class Data_Add(Student):
    def __init__(self,roll_no,name,age,course):
        super().__init__(roll_no,name,age)   # super internally : Student.__init__(self,roll_no,name,age)
        self.course = course
        
    def show_course(self):
        super().show()
        print("course : ",self.course)
        
add = Data_Add(2,"Akash 2",16,"BCA")
add.show_course()


print("hello")



# class Model:
#     def __init__(self):
#         self.weights = [0.2,0.5]
    
#     def predict(self,x):
#         return x * self.weights[1]

# model = Model()
# model.predict(1000)


```
