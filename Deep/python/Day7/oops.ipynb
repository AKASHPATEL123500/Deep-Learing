# Oops object oriented programing


*Problem 1: create a object class*

```python
class Car:
    def __init__(self,name,username):
        self.name = name
        self.username = username
    
    # full name print
    def full_name(self):
        return f"{self.name} {self.username}"
    
my_car = Car("akash","akash_reddy_01")
print(my_car.name)
print(my_car.username)
print(my_car.full_name())

my_new_car = Car("akash","akash_reddy_01")
print(my_new_car.name)


print("\n")


# inerhtence
class Electric_Car(Car):
    def __init__(self,name,username,age):
        super().__init__(name,username)
        self.age = age

new_card_added_tesla = Electric_Car("akash","akash_reddy_01",17)
print(new_card_added_tesla)
```
