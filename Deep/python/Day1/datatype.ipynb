# Example: Modifying and Accessing List Elements
The Python code above defines three lists (`number`, `string` and `mixed`) and
demonstrates basic list operations: printing the list, changing an element
by index, and iterating over elements with a `for` loop.  
We also show how to access individual items using their index
(e.g. `number[0] = 22` and `print(string[0])`).  
The loop uses the variable `i` (currently 3) to hold each index as it runs.
These examples illustrate that lists are mutable (their contents can be
changed) and can contain different data types.

```python
number = [10,20,30,40]
print(number)
number[0]=22
print(number)

string = ["a","b","c","d"]
print(string)
print(string[0])

for i in range(len(string)):
    print(i)

mixed = [True,"name",12,5.05]
print(mixed)
print(mixed[0])
```
