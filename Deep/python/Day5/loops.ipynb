# loops

```python
for i in range(4):
    print("hello world")

for i in range(0,4):
    print(i)

```

# print even numbers

```python
for i in range(1,11):
    if i % 2 == 0:
        print(i)
```

# print odd numbers

```python
for i in range(1,11):
    if i % 2 == 1:
        print(i)
```

# While loop

```python
i = 1
while i <= 5:
    print(i)
    i=i+1
```

# Nested Loop

```python
for i in range(3):
    for j in range(1):
        print(i,j)
```

```python
# problem : find logest continous strictly increasing sub array

arr = [12,4,7,9,2,8,10,12,4]

n = len(arr)

current_len = 1
max_len = 1

current_start = 0
max_start = 0

for i in range(1,n):
    if arr[i] > arr[i-1]:
        current_len += 1
    else:
        if current_len > max_len:
            max_len = current_len
            max_start = current_start
        current_len = 1
        current_start = 1

if current_len > max_len:
    max_len = current_len
    max_start = current_start   

result = arr[max_start:max_start + max_len]
print("length of array: ", max_len)
print("result : ",result)
```
