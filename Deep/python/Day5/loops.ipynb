# loops

```python
for i in range(4):
    print("hello world")

for i in range(0,4):
    print(i)

```

# print even numbers

```python
for i in range(1,11):
    if i % 2 == 0:
        print(i)
```

# print odd numbers

```python
for i in range(1,11):
    if i % 2 == 1:
        print(i)
```

# While loop

```python
i = 1
while i <= 5:
    print(i)
    i=i+1
```

# Nested Loop

```python
for i in range(3):
    for j in range(1):
        print(i,j)
```
