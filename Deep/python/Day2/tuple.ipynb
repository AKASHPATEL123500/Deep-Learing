# Tuple  = a tuple is a collection of values that can not be changed

`Tuple = fixed value`

*Syntax:*

`a = (1,2,3,4,5)`


# check tuple is muteable or not 

```python
nums = (1,2,3,4,5)
nums[0] = 20
print(nums)


# tuple : is not muteable 
```

```python
nums = (1,2,3,4,5)
print(nums)
```

# Negative indexing

```python
num = (1,2,3,4,5)
print(num[-1])
print(num[-2])
print(num[-3])
print(num[-4])
```

# Tuple slicing

```python
num = (1,2,3,4,5)
print(num[:1])
print(num[:2])
print(num[:3])
print(num[:4])
print(num[:5])
print("\n")
print(num[4:])
print(num[3:])
print(num[2:])

```

# Loop through access

```python
num=(1,2,3,4,5,5)
for i in num:
    print(i)
```

```python
num=(1,2,3,4,5,5)
for i in range(len(num)):
    print(num[i])
```

# Tuple functions

*1: `Count()`*


```python
num = (1,2,3,4,2,3,4,44,4,4,4,4,4,4,4,4,4,4,4)
counnts = num.count(4)
print(counnts)
```

*2: `Index()`*

```python
num = (1,2,3,4,2,3,4,44,4,4,4,4,4,4,4,4,4,4,4)
print(num.index(3))
```

*2: `Legth()`*

```python
num = (1,2,3,4,2,3,4,44,4,4,4,4,4,4,4,4,4,4,4)
print(len(num))
```

```python
a = (1)
b = (1,)
print(a)
print(b)
```
