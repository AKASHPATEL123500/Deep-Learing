*problem 4: Create a function that retunrs both the area and circumference of a circle given its radius*

```python
import math

def circle(radius):
    area = math.pi * radius ** 2   # radius * radius
    circumference = 2 * math.pi * radius
    return area,circumference

a , c = circle(2)
# print("Area :",round(a,2), "\n" ,"circumference :",round(c,2))
print(f"Area: {a:.2f}")
print(f"circumference: {c:.2f}")
```

*Problem 7: Write a function that take a variable numbers of agrumnets and returns their sum*

`*args`

```python
def sum_all(*args):
    print(args)
    for i in args:
        print( i * 2 )
    return sum(args)  
print(sum_all(1,2,3,4))
```

*Problem 8: funtion with `**args`*

```python
def print_kwargs(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_kwargs(
    name = "Akash",
    username = "akash_reddy_01",
    age = 17
)

print_kwargs(
    name = "Akash",
    username = "akash_reddy_01",
    age = 17,
    city = "prayagraj",
    pincode = 221503
)

print_kwargs(
    name = "Akash",
    username = "akash_reddy_01",
)



```

*problem 9 : Genrateor function with yeild*

```python
def even_genrator(limit):
    for i in range(2,limit + 1 , 2):
        yield i

# yeild value ko return karta hai aur memory me usko store bhi karta hai 

for num in even_genrator(10):
    print(num)
```

*Problem 10 : Recursice funtion*

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)
    

print(factorial(2))
```
