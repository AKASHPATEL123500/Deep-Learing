# Dictonary
`dictonary is a data type that store data in key paires`

`Example:`

```python
my_dict ={
    "name" : "akash",
    "username" : "akash_reddy_01",
    "marks": 85,
    "is_passed" : True
}
```

# Accessing dictonary value

```python
my_dict ={
    "name" : "akash",
    "username" : "akash_reddy_01",
    "marks": 85,
    "is_passed" : True
}

print(my_dict["name"])
print(my_dict["username"])
print(my_dict["marks"])
print(my_dict["is_passed"])
```

# Accesing dict value using method 2 `get()`

```python
my_dict ={
    "name" : "akash",
    "username" : "akash_reddy_01",
    "marks": 85,
    "is_passed" : True
}

print(my_dict.get("name"))
print(my_dict.get("username"))
print(my_dict.get("marks"))
print(my_dict.get("is_passed"))
```

# Adding value

```python
data = {
    "city":"allahabad",
    "pincode":221503,
    "state":"uttar perdesh"
}


data["city"]= "prayagraj"
print(data.get("city"))

data.pop("city")
print(data)

del data["pincode"]
print(data)


data.clear()
print(data)


```

# Accessing dict through loop

```python
students = {
    "name":"akash",
    "username":"akash_reddy",
    "city":"prayagraj",
    "pincode":221503,
    "mobile_no": 1234567890
}

# print key
for key in students:
    print(key)

print("\n")

# print value
for value in students.values():
    print(value)

print("\n")


# print key + value
for key, value in students.items():
    print(key,value)

```
