```python
class add:
    def __init__(self, a, b):
        self.a = a
        self.b = b

    def calculate(self):
        return self.a + self.b
add1 = add(1, 2)
print(add1.calculate())
```

```python
arr1 = [1,2,3,4,5]
arr2 = [3,4,5,6,7]

result = []

for i in range(len(arr1)):
    for j in range(len(arr2)):
        if arr1[i] == arr2[j]:
            if arr1[i] not in result:
                result.append(arr1[i])
            print(result)
```
